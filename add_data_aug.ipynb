```python
import pandas as pd
import os
import shutil
from pprint import pprint
```

```python
start_num = 2943
new_data = []
```

```python
to_add_data = pd.read_csv('Data/final_data_2.csv')
extra_wavs = os.listdir('WavFormat/andthis/')
```

```python
to_add_data.pop('Unnamed: 0')
to_add_data
```

```python
for file in extra_wavs:
    info = {'speakerid': start_num, 'filename': 1, 'birthplace': 2, 'native_language': 3, 'sex': 4}
    start_num += 1
    file_name = file.split('_')[0]
    row_num = to_add_data[to_add_data['filename'] == file_name].index[0]
    info['filename'] = os.path.splitext(file)[0]
    info['birthplace'] = to_add_data.loc[row_num].birthplace
    info['native_language'] = to_add_data.loc[row_num].native_language
    info['sex'] = to_add_data.loc[row_num].sex
    new_data.append(info)
    shutil.copy(f'WavFormat/andthis2/{file}',f'WavFormat/{file}')
```

```python
df = pd.DataFrame(new_data)
df
```

```python
final_data = pd.concat([to_add_data, df])
final_data
```

```python
final_data.to_csv('Data/3_lang.csv')
```
